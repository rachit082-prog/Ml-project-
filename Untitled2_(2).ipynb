```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
dataset = pd.read_csv("weather.csv")
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().any()
```

```python
# Drop rows with null values
dataset_cleaned = dataset.dropna()

# Drop columns with null values (if entire column is irrelevant)
dataset_cleaned = dataset.dropna(axis=1)
```

```python
temp = dataset['temperaturemin'].values.reshape(-1, 1)  # Independent variable
precipitation = dataset['precipitation'].values  # Dependent variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(temp, precipitation, test_size=1/3, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print(f"Coefficient (m): {regressor.coef_[0]}")
print(f"Intercept (c): {regressor.intercept_}")
```

```python
y_train_pred = regressor.predict(x_train)
y_test_pred = regressor.predict(x_test)
```

```python
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
```

```python
print(f"Training RMSE: {rmse_train:.2f}")
print(f"Testing RMSE: {rmse_test:.2f}")
```

```python
plt.scatter(x_train, y_train, color='green', label="Actual")
plt.plot(x_train, y_train_pred, color='black', label="Predicted")
plt.title("Training Dataset: Temperature vs Precipitation")
plt.xlabel("Temperature (°C)")
plt.ylabel("Precipitation (mm)")
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue', label="Actual")
plt.plot(x_train, regressor.predict(x_train), color='black', label="Model Fit")
plt.title("Testing Dataset: Temperature vs Precipitation")
plt.xlabel("Temperature (°C)")
plt.ylabel("Precipitation (mm)")
plt.legend()
plt.show()
```

```python
new_temperatures = np.array([5, 15, 25]).reshape(-1, 1)  # Example temperatures
predicted_precipitation = regressor.predict(new_temperatures)

# Display the Predictions
for temp, precip in zip(new_temperatures.flatten(), predicted_precipitation):
    print(f"At {temp}°C, the predicted precipitation is {precip:.2f} mm.")
```

```python
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['month'] = dataset['date'].dt.month
dataset['day'] = dataset['date'].dt.day
dataset['year'] = dataset['date'].dt.year
```

```python
yes_no_columns = [
    'fog', 'fogheavy', 'mist', 'rain', 'fogground', 'ice', 'glaze',
    'drizzle', 'snow', 'freezingrain', 'smokehaze', 'thunder',
    'highwind', 'hail', 'blowingsnow', 'dust', 'freezingfog'
]
for col in yes_no_columns:
  dataset[col] = dataset[col].map({'Yes': 1, 'No': 0})
```

```python
dataset.fillna(dataset.mean(), inplace=True)
```

```python
X_temp = dataset[['month', 'day', 'year', 'precipitation', 'avgwindspeed']]
y_temp_min = dataset['temperaturemin']
y_temp_max = dataset['temperaturemax']
```

```python
X_temp_train, X_temp_test, y_temp_min_train, y_temp_min_test = train_test_split(
    X_temp, y_temp_min, test_size=0.2, random_state=42
)
_, _, y_temp_max_train, y_temp_max_test = train_test_split(
    X_temp, y_temp_max, test_size=0.2, random_state=42
)
```

```python
temp_model_min = RandomForestRegressor(random_state=42)
temp_model_max = RandomForestRegressor(random_state=42)
temp_model_min.fit(X_temp_train, y_temp_min_train)
temp_model_max.fit(X_temp_train, y_temp_max_train)
```

```python
y_temp_min_pred = temp_model_min.predict(X_temp_test)
y_temp_max_pred = temp_model_max.predict(X_temp_test)
print("Temperature Range Prediction:")
print(f"Min Temperature RMSE: {np.sqrt(mean_squared_error(y_temp_min_test, y_temp_min_pred))}")
print(f"Max Temperature RMSE: {np.sqrt(mean_squared_error(y_temp_max_test, y_temp_max_pred))}")
```

```python
X_rain = dataset[['month', 'day', 'year', 'temperaturemin', 'temperaturemax', 'avgwindspeed']]
y_rain = dataset['rain']
```

```python
X_rain_train, X_rain_test, y_rain_train, y_rain_test = train_test_split(
    X_rain, y_rain, test_size=0.2, random_state=42
)
```

```python
rain_model = RandomForestClassifier(random_state=42)
rain_model.fit(X_rain_train, y_rain_train)
```

```python
y_rain_pred = rain_model.predict(X_rain_test)

print("\nRainfall Likelihood Prediction:")
print(f"Accuracy: {accuracy_score(y_rain_test, y_rain_pred)}")
```
